```python
# 1.  dealing with missing values
#missing values represent by ?,N/A,0,  or just blank cell

#deal with missing values or data
# a.  check the data source
# b.  drop the missing values(drop the data entery(drop the whole row))
# or  Replace the missing vlues(by the average of the whole column or  replace by maximum frequent number)      this is best way   
# d   leave that as missing
```

```python
# b  how we drop the missing values
# for this we use the dropna() function of pandas(this function drop all row and column that are missing)
df.dropna()
```

```python
#if we set the axis=0 the drop the whole row that element is missing
#if we set axis=1 that drop the whole column(this is dangerous for result)

df1=df.dropna(subset=["ca"],axis=0,inplace=True)
```

```python
# replace the missing values
#first we take the mean of that column
mean=df["ca"].mean()
mean
#replace by the replace function they autimatically find NAN values
rep=df["ca"].replace(to_replace ="NaN",   value =mean)
rep
```

```python
# convert the (percent_of_housing_crowded) datatyoe float into int by using astype() function
df["percent_of_housing_crowded"]=df["percent_of_housing_crowded"].astype("int")
df["percent_of_housing_crowded"]
```

```python

# 3.  data normaliztion(centering,scaling) #data mining me kiyia tha
# data normalization(uniform the feature value between different ranges) is important before we do the data analization
# basically ye is tra ha k hr column ki value dosery column k increasee ho asa na ho aik column ki value increase ho ri ha or dosery ki decrease ho ri ho
#ku k aik hsb sy sari aik tarteb me a ri ho to compare krna easy ho jy ga or data analisation b best ho gi
# ku k aik hsb sy bsari increase ya decrease ho ri ho to predication krna easy ho jta ha

    #methods for normalisation
    # 1 simple feature scaling
    # 2 Min_Max
    # 3 Z score(standard score)
    
#Example(1 simple feature scaling)
df["ca"]=df["ca"] / df["ca"].max()
df["ca"]


```

```python
#Example(2 Min_Max)
df["ca"]=(df["ca"] -  df["ca"].min())   /  (df["ca"].max() -  df["ca"].min()) 
df["ca"]

#3nu method ka purpose samwe ha
```

```python
#Example(3 Z_score)(standard method)
df["ca"]=(df["ca"] -  df["ca"].mean())   /  df["ca"].std()
df["ca"]
```
